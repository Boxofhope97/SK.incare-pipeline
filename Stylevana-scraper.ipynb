```python
import requests
import json
from bs4 import BeautifulSoup

for x in range(1,31):
    url = f'https://www.stylevana.com/en_US/best-sellers.html?pg={x}'
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'lxml')
    productlist = soup.find_all('div', class_='product-item-details')

    for item in productlist:
    # Extract brand, name, current price, original price, discount percentage
        name = item.find('h2', class_='product-name').text.strip()
        current_price = item.find('span', class_='price').text.strip()
        original_price = item.find('span', class_='price').text.strip()
        discount_percent = item.find('span', class_='percent').text.strip()

        # store the information in a data structure: list and append each item accordingly.
        all_products = []
        product_info = {"Brand + Name": name, "current_price": current_price, "original_price": original_price, "discount_percent": discount_percent}
        all_products.append(product_info)
        print(all_products)
        # Print or store the information as needed
        # print(f"{brand}, {name}, {price},{sale_price},{discount_percent}")
```

```python
# Create csv and write product list to output file
import csv

with open('Stylevana-bestselling-products.csv','w', newline='') as f_output:
    csv_output = csv.writer(f_output)
    csv_output.writerows(all_products)
```
